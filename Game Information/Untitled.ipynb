```r
# Load necessary libraries
library(ggplot2)
library(gganimate)
library(magick)

# Create sample dataset
df <- data.frame(
    time = rep(1:10, each = 3),
    product = rep(letters[1:3], 10),
    price = runif(30, min = 10, max = 50)
)

# Generate the animated bar chart
p <- ggplot(df, aes(x = product, y = price, fill = product)) +
    geom_bar(stat = "identity") +
    labs(title = 'Time: {frame_time}', y = 'Price', x = 'Product') +
    transition_time(time) +
    ease_aes('linear') +
    theme_minimal()

# Animate and render the plot in memory
anim <- animate(p, nframes = 100, fps = 10, renderer = magick_renderer())

# Display the animation directly in RStudio
anim
```
